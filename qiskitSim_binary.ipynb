```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
num_bits = 5
circuit = QuantumCircuit(num_bits, num_bits)
```

```python
# Apply Hadamard gate to put qubits in superposition
for qubit in range(num_bits):
    circuit.h(qubit)

# Measure qubits
circuit.measure(range(num_bits), range(num_bits))
```

```python
# Initialize the AerSimulator
simulator = AerSimulator()

# Transpile the circuit for the simulator
from qiskit import transpile
compiled_circuit = transpile(circuit, simulator)
```

```python
# Run the circuit on the simulator
job = simulator.run(compiled_circuit, shots=20000)

# Get the results
result = job.result()
counts = result.get_counts(compiled_circuit)
```

```python
# Extract the random bit strings from the counts
random_bits = []
for key in counts:
    random_bits.extend([key] * counts[key])  # Extend the list with repeated outcomes

# Combine into a single string
combined_bits = ''.join(random_bits)
print(f"Total random bits generated: {len(combined_bits)}")
print(f"First 100000 bits: {combined_bits[:100000]}")
```
